```python
import spacy

#load the spacy model 
nlp = spacy.load("en_core_web_sm")
#input String 
input_str = "Hello$ who# are@ yo!u" 

#Function to clean the data 
def clean_text(text): 
    clean_text = ''.join(char for char in text if char.isalpha() or char.isspace())
    doc = nlp(clean_text)
    return ' '.join(token.text for token in doc)

#Get the final output 
clean_str = clean_text(input_str)
print(clean_str)
```

```python
import re

def expand_contractions(text): 
    contractions_pattern = {
        r"(?i)can't": "cannot",
        r"(?i)won't": "will not",
        r"(?i)it's": "it is",
        r"(?i)weren't": "were not",
        r"(?i)i'm": "i am"
    }

    for contraction, expansion in contractions_pattern.items(): 
        text = re.sub(contraction, expansion, text)
    
    return text

txt = "I can't brlieve it's already running" 
expanded_txt = expand_contractions(txt)
print(expanded_txt)
```

n - Grams

```python
import nltk 
from nltk.tokenize import word_tokenize 
from nltk import ngrams 

def generate_ngrams(text, n):
    tokens = word_tokenize(txt)
    n_grams = list(ngrams(tokens, n))
    return n_grams
#Example text 
txt = "N grams are a sequence of n items for a given sample of text or speech"

unigrams = generate_ngrams(txt, 1)
bigrams = generate_ngrams(txt, 2) 
trigrams = generate_ngrams(txt, 3)

```

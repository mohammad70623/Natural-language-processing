Libraries in the field of NLP 
1. spaCy - NLP libraries in the pyhon that can be used to tokenize and process texual data
2. nltk

```python
import spacy

#load the spacy model 
nlp = spacy.load("en_core_web_sm")
#input String 
input_str = "Hello$ who# are@ yo!u" 

#Function to clean the data 
def clean_text(text): 
    clean_text = ''.join(char for char in text if char.isalpha() or char.isspace())
    doc = nlp(clean_text)
    return ' '.join(token.text for token in doc)

#Get the final output 
clean_str = clean_text(input_str)
print(clean_str)
```

```python
# from nltk.tokenize import word_tokenize
input_str = "Hello$ who# are@ yo!u"  
#Tokenize 
tokens = nltk.word_tokenize(input_str)

#Remove the special character 

clean_tokens = [token for token in tokens if token.isalnum() ]

clean_str = ' '.join(clean_tokens)
print(clean_str)
```

```python
import re

def expand_contractions(text): 
    contractions_pattern = {
        r"(?i)can't": "cannot",
        r"(?i)won't": "will not",
        r"(?i)it's": "it is",
        r"(?i)weren't": "were not",
        r"(?i)i'm": "i am"
    }

    for contraction, expansion in contractions_pattern.items(): 
        text = re.sub(contraction, expansion, text)
    
    return text

txt = "I can't brlieve it's already running" 
expanded_txt = expand_contractions(txt)
print(expanded_txt)
```

Tokenization

```python
import nltk 
from nltk.tokenize import word_tokenize, sent_tokenize 

#Sample text for tokenization 
txt = "NLTK provides powerful tools for tokenization. It include word tokenization and sentence tokenzation"
#Word tokenization 
words = word_tokenize(txt)
print(words)
```

```python
#sentence tokenization 

sentance = sent_tokenize(txt)
print(sentance)
```
